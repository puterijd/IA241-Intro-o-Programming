# COVID-19

COVID-19 is a highly infectious virus spreading across the world causing extreme problems for many countries. The virus is putting extreme stress on the infrastructure of all of these countries and leading to hundreds of thousands of deaths around the globe. The following is analysis of data concerning the COVID-19 Virus and specifically the case and death numbers associated with it.

The information was retrieved from the European Center for Disease Control on April 28, updated information is available on the website:https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide

```python
%matplotlib inline
import pandas
```

# Entirety of the Data Imported from S3

```python
df = pandas.read_excel('s3://puterio-ia241/COVID-19-geographic-disbtribution-worldwide-2020-04-28.xlsx')
df
```

# Number of Deaths Per Month

```python
sum_monthly_deaths_total = df.groupby('month').sum()['deaths']

sum_monthly_deaths_total.plot.bar()
```

The Graph shows a huge spike in month 3 (March) and a continued increase in month 4 (April)

# The Max Number of Deaths in a Day

```python
import numpy as np
```

```python
print('the max deaths in a day is {}'.format(np.amax(df['deaths'])))
```

```python
print('the max cases in a day is {}'.format(np.amax(df['cases'])))
```

Shows that the maximum Number of deaths in a single day for the world was 4928 while the max number of cases was 48529.

# France Data From Past 30 Days

```python
france_data = df.loc[df['countriesAndTerritories'] == 'France']
france_data[:30] #the top 30 rows
```

```python
france_data.plot(x='dateRep',y='cases')
```

This Chart shows the spike in cases in france between March 30 2020 and April 28 2020. The genaral trend was an increase but cases saw a small decline in the final 10 to 15 days.

# Countries With Smallest Number of Cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_by_country.nsmallest(20).plot.bar()
```

The chart shows the 20 countries with the smallest amount of cases, based on the data, every country has been infected by at least 1 case.

# Global Fatality Rate

```python
global_cases = df.sum()['cases']
print(global_cases)
```

```python
global_deaths = df.sum()['deaths']
print(global_deaths)
```

```python
global_fatality_rate = (global_deaths)/(global_cases)
print(global_fatality_rate)
```

The charted information has a global fatality rate of just over 7 percent.
